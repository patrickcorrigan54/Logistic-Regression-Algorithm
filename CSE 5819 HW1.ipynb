## Exercise 1-1
### An equivalent expression to (x$^T$AB)$^T$ would be (AB)$^T$x. This is due to the fact that (AB)$^T$ = B$^T$A$^T$ . In the original equation (x$^T$AB)$^T$, if the product of AB is treated as a single matrix, then using the fact stated above, we can rewrite this expression to be (AB)$^T$(x$^T$)$^T$, which simplifies to be (AB)$^T$x. This is demonstrated with random matrices in the code below. 

```python
n = 5
x = np.random.randint(0,5, (n,1))
A = np.random.randint(0,5, (n,n))
B = np.random.randint(0,5, (n,n))

xt = x.T
AB = np.dot(A, B)
xtAB = np.dot(xt, AB)
xtABt = xtAB.T
print('(x^t * AB)^t = ')
print(xtABt)
print('')

ABt = AB.T
ABtx = np.dot(ABt, x) 
print('(AB)^t * x = ')
print(ABtx)
```

### The plot with different $\theta$ values can be seen below. The $\theta$ value of 1 is plotted in black, the $\theta$ value of 2 is plotted in blue, and the $\theta$ value of 5 is plotted in red.

### As $\theta$ increases, the function $g$ has a faster transition from a y value of 0 to 1. That is to say, the slope of the function near the x value of 0 increases as the value of theta increases, and the function has a sharper, less gradual shape. 

### As $\theta$ approaches infinity, the function $g$ would become an almost instant transition from a y value of 0 to 1 at an x value of 0. That is to say, the function would be flat at all points not at an x value of 0, and would move instantly from a y value of 0 to 1 at an x value of 0. 

```python
def sigmoid( x ):
    g = 1 / (1 + math.exp(-x))
    return g
vsigmoid = np.vectorize(sigmoid)
theta1 = 1
theta2 = 2
theta5 = 5
x = np.arange(-4, 4, .1)
output_theta1x = vsigmoid(theta1*x)
output_theta2x = vsigmoid(theta2*x)
output_theta5x = vsigmoid(theta5*x)

plt.plot(x, output_theta1x, 'black');
plt.plot(x, output_theta2x, 'blue');
plt.plot(x, output_theta5x, 'red');
```
